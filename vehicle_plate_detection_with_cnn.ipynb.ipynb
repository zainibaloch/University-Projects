```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tqdm import tqdm

# Step 1: Load Pretrained CNN (VGG16)
model = VGG16(weights='imagenet', include_top=False, pooling='avg')

# Step 2: Extract Features from Images
def extract_features(image_folder):
    features = []
    labels = []

    # Iterate over class folders
    for label_folder in os.listdir(image_folder):
        label_path = os.path.join(image_folder, label_folder)
        if not os.path.isdir(label_path):
            continue  # Skip if not a directory

        # Iterate over images in the folder
        for image_file in tqdm(os.listdir(label_path), desc=f"Processing {label_folder}"):
            image_path = os.path.join(label_path, image_file)

            try:
                # Load and preprocess the image
                image = load_img(image_path, target_size=(224, 224))  # Resize for VGG16
                image_array = img_to_array(image)
                image_array = np.expand_dims(image_array, axis=0)  # Add batch dimension
                image_array = preprocess_input(image_array)  # Normalize for VGG16

                # Extract features
                feature = model.predict(image_array)
                features.append(feature.flatten())  # Flatten the feature map
                labels.append(label_folder)  # Store the label (folder name as class)
            except Exception as e:
                print(f"Error processing {image_path}: {e}")

    return np.array(features), labels

# Step 3: Save Features and Labels to CSV
def save_to_csv(features, labels, output_file):
    """Save features and labels to a CSV file."""
    df = pd.DataFrame(features)
    df['Label'] = labels
    df.to_csv(output_file, index=False)

# Paths
image_folder = "C:/Users/User/Desktop/train"  # Path to the train folder
output_csv = "mri_features.csv"  # Output CSV file

# Process Images and Save to CSV
features, labels = extract_features(image_folder)
save_to_csv(features, labels, output_csv)

print(f"Features saved to {output_csv}")
```

```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input

# Load pre-trained VGG16 model + top (classification) layers excluded
model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Step 2: Extract Features from Images
def extract_features(image_folder):
    features = []
    labels = []

    # Iterate over class folders
    for label_folder in os.listdir(image_folder):
        label_path = os.path.join(image_folder, label_folder)
        if not os.path.isdir(label_path):
            continue  # Skip if not a directory

        # Iterate over images in the folder
        for image_file in tqdm(os.listdir(label_path), desc=f"Processing {label_folder}"):
            image_path = os.path.join(label_path, image_file)

            try:
                # Load and preprocess the image
                image = load_img(image_path, target_size=(224, 224))  # Resize for VGG16
                image_array = img_to_array(image)
                image_array = np.expand_dims(image_array, axis=0)  # Add batch dimension
                image_array = preprocess_input(image_array)  # Normalize for VGG16

                # Extract features using the VGG16 model
                feature = model.predict(image_array)
                feature = feature.flatten()  # Flatten the feature map
                features.append(feature)  # Add flattened feature to list
                labels.append(label_folder)  # Store the label (folder name as class)
            except Exception as e:
                print(f"Error processing {image_path}: {e}")

    return np.array(features), labels

# Step 3: Save Features and Labels to CSV
def save_to_csv(features, labels, output_file):
    """Save features and labels to a CSV file."""
    df = pd.DataFrame(features)
    df['Label'] = labels
    df.to_csv(output_file, index=False)

# Paths
image_folder = "C:/Users/User/Desktop/test"  # Path to the test folder
output_csv = "mri_features_test.csv"  # Output CSV file

# Process Images and Save to CSV
features, labels = extract_features(image_folder)
save_to_csv(features, labels, output_csv)

print(f"Features saved to {output_csv}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


# Ensure consistent columns between training and testing datasets
common_columns = train_data.columns.intersection(test_data.columns)
train_data = train_data[common_columns]
test_data = test_data[common_columns]

# Split features and labels
X_train = train_data.iloc[:, :-1].values  # All columns except the last one (features)
y_train = train_data['Label'].values      # Last column (labels)

X_test = test_data.iloc[:, :-1].values
y_test = test_data['Label'].values

# Encode labels
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Logistic Regression
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)

# Train SVM
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)




# Calculate Training and Testing Accuracy
lr_train_acc = accuracy_score(y_train, lr_model.predict(X_train))
lr_test_acc = accuracy_score(y_test, lr_model.predict(X_test))

svm_train_acc = accuracy_score(y_train, svm_model.predict(X_train))
svm_test_acc = accuracy_score(y_test, svm_model.predict(X_test))

# Evaluate Logistic Regression
lr_pred = lr_model.predict(X_test)
print("Logistic Regression Model")
print("Accuracy:", accuracy_score(y_test, lr_pred))
print("\nClassification Report:\n", classification_report(y_test, lr_pred, target_names=label_encoder.classes_))

# Evaluate SVM
svm_pred = svm_model.predict(X_test)
print("\nSupport Vector Machine Model")
print("Accuracy:", accuracy_score(y_test, svm_pred))
print("\nClassification Report:\n", classification_report(y_test, svm_pred, target_names=label_encoder.classes_))



# Create a bar chart for visualization
models = ['Logistic Regression', 'SVM']
train_accuracies = [lr_train_acc, svm_train_acc]
test_accuracies = [lr_test_acc, svm_test_acc]

x = range(len(models))  # X-axis positions

plt.figure(figsize=(8, 5))
plt.bar(x, train_accuracies, width=0.4, label='Training Accuracy', color='blue', align='center')
plt.bar([i + 0.4 for i in x], test_accuracies, width=0.4, label='Testing Accuracy', color='orange', align='center')

# Add labels, title, and legend
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Training vs. Testing Accuracy')
plt.xticks([i + 0.2 for i in x], models)  # Set X-axis labels at the center of bars
plt.ylim(0, 1)  # Accuracy is between 0 and 1
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

# Load datasets
train_file = "C:/Users/User/Desktop/mri_features.csv"
test_file = "C:/Users/User/Desktop/mri_features_testing.csv"
train_data = pd.read_csv(train_file)
test_data = pd.read_csv(test_file)

# Ensure consistent columns between training and testing datasets
common_columns = train_data.columns.intersection(test_data.columns)
train_data = train_data[common_columns]
test_data = test_data[common_columns]

# Split features and labels
X_train = train_data.iloc[:, :-1].values  # All columns except the last one (features)
y_train = train_data['Label'].values      # Last column (labels)

X_test = test_data.iloc[:, :-1].values
y_test = test_data['Label'].values

# Encode labels
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize models
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train models
xgb_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)

# Predict and evaluate
models = {
    'XGBoost': xgb_model,
    'Random Forest': rf_model,
    'KNN': knn_model
}

train_accuracies = []
test_accuracies = []

for model_name, model in models.items():
    # Training accuracy
    train_acc = accuracy_score(y_train, model.predict(X_train))
    train_accuracies.append(train_acc)

    # Testing accuracy
    test_acc = accuracy_score(y_test, model.predict(X_test))
    test_accuracies.append(test_acc)

    # Classification report
    print(f"--- {model_name} ---")
    print(f"Training Accuracy: {train_acc:.2f}")
    print(f"Testing Accuracy: {test_acc:.2f}")
    print("Classification Report:")
    print(classification_report(y_test, model.predict(X_test), target_names=label_encoder.classes_))
    print()

# Plot Training vs. Testing Accuracy
x = range(len(models))  # X-axis positions

plt.figure(figsize=(10, 6))
plt.bar(x, train_accuracies, width=0.4, label='Training Accuracy', color='blue', align='center')
plt.bar([i + 0.4 for i in x], test_accuracies, width=0.4, label='Testing Accuracy', color='orange', align='center')

# Add labels, title, and legend
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Training vs. Testing Accuracy for XGBoost, Random Forest, and KNN')
plt.xticks([i + 0.2 for i in x], models.keys())  # Set X-axis labels at the center of bars
plt.ylim(0, 1)  # Accuracy is between 0 and 1
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
